## Bài 1 - Can Chi Calendar

Viết hàm `calculate_can_chi_calendar(year)` để kiểm tra một năm bất kỳ là năm Can Chi gì.

```python
def calculate_can_chi_calendar(year):
    """Return Can Chi name for a Gregorian year."""
    if not isinstance(year, int):
        raise TypeError("year must be an integer")

    can = ["Canh", "Tân", "Nhâm", "Quý", "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
    chi = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu", "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]
    return f"{can[year % 10]} {chi[year % 12]}"


for test_year in [2025, 2024, 1997]:
    print(f"calculate_can_chi_calendar({test_year}) -> {calculate_can_chi_calendar(test_year)}")
```

## Bài 2 - Tính giá trị tài khoản đầu tư

Tính tổng giá trị tài khoản sau `T` năm với góp tiền đều hàng tháng, lãi kép theo tháng và lãi suất thay đổi theo từng năm.

Giả định tính toán: cuối mỗi tháng tài khoản nhận lãi trước, sau đó nhà đầu tư góp thêm `PMT`.

```python
def calculate_investment_value(P, PMT, T, annual_rates):
    """Calculate future value with monthly compounding and yearly variable rates.

    annual_rates can be passed as percentages, e.g. 7 for 7%, or decimals, e.g. 0.07.
    Contributions are added at the end of each month.
    """
    if T <= 0:
        raise ValueError("T must be positive")
    if len(annual_rates) != T:
        raise ValueError("annual_rates must contain exactly T rates")

    balance = float(P)
    yearly_summary = []

    for year_index, annual_rate in enumerate(annual_rates, start=1):
        normalized_rate = annual_rate / 100 if annual_rate > 1 else annual_rate
        monthly_rate = normalized_rate / 12

        for _ in range(12):
            balance *= 1 + monthly_rate
            balance += PMT

        yearly_summary.append({
            "year": year_index,
            "annual_rate": annual_rate,
            "balance": balance,
        })

    return balance, yearly_summary


P = 100_000_000
PMT = 2_000_000
T = 5
annual_rates = [7, 7.5, 7.8, 8, 8.5]

final_value, yearly_summary = calculate_investment_value(P, PMT, T, annual_rates)
for item in yearly_summary:
    print(f"Năm {item['year']} - lãi suất {item['annual_rate']}%: {item['balance']:,.0f} VND")
print(f"Tổng giá trị tài khoản sau {T} năm: {final_value:,.0f} VND")
```

## Bài 3 - Random số thỏa điều kiện

Chọn ngẫu nhiên hai số `x` và `y` trong `[1, 20]`. Hàm `random_number_with_condition(target_sum)` trả về số lần random cần thực hiện cho đến khi `x + y == target_sum`.

Trong ví dụ demo dùng `seed=0` để kết quả ổn định khi chạy lại notebook.

```python
import random


def random_number_with_condition(target_sum, lower=1, upper=20, seed=None, max_attempts=1_000_000):
    """Random pairs until x + y equals target_sum, then return attempt count and pair."""
    min_sum = lower + lower
    max_sum = upper + upper
    if not (min_sum <= target_sum <= max_sum):
        raise ValueError(f"target_sum must be between {min_sum} and {max_sum}")

    rng = random.Random(seed)
    for attempt in range(1, max_attempts + 1):
        x = rng.randint(lower, upper)
        y = rng.randint(lower, upper)
        if x + y == target_sum:
            return attempt, (x, y)

    raise RuntimeError("Condition was not met within max_attempts")


attempt_count, matched_pair = random_number_with_condition(30, seed=0)
print(f"random_number_with_condition(40) -> {attempt_count} lần")
print(f"Cặp số thỏa điều kiện cuối cùng: {matched_pair[0]} + {matched_pair[1]} = 40")
```

## Bài 4 - Xây dựng class Stack

Xây dựng class `MyStack` có `capacity` và các method: `.isEmpty()`, `.isFull()`, `.pop()`, `.push(value)`, `.top()`.

```python
class MyStack:
    def __init__(self, capacity):
        if not isinstance(capacity, int):
            raise TypeError("capacity must be an integer")
        if capacity <= 0:
            raise ValueError("capacity must be positive")
        self.capacity = capacity
        self._items = []

    def isEmpty(self):
        return len(self._items) == 0

    def isFull(self):
        return len(self._items) == self.capacity

    def push(self, value):
        if self.isFull():
            raise OverflowError("Stack is full")
        self._items.append(value)

    def pop(self):
        if self.isEmpty():
            raise IndexError("Stack is empty")
        return self._items.pop()

    def top(self):
        if self.isEmpty():
            raise IndexError("Stack is empty")
        return self._items[-1]

    def __repr__(self):
        return f"MyStack(capacity={self.capacity}, items={self._items})"


stack1 = MyStack(capacity=5)
stack1.push(1)
stack1.push(2)

print("Ví dụ trong đề:")
print(f"stack1.isFull() -> {stack1.isFull()}")
print(f"stack1.top() -> {stack1.top()}")
print(f"stack1.pop() -> {stack1.pop()}")
print(f"stack1.top() -> {stack1.top()}")
print(f"stack1.pop() -> {stack1.pop()}")
print(f"stack1.isEmpty() -> {stack1.isEmpty()}")

print("\nDemo capacity = 4 theo hình minh họa:")
stack2 = MyStack(capacity=4)
for value in [31, 12, 10, 20]:
    stack2.push(value)
    print(f"push({value}) -> top = {stack2.top()}, stack = {stack2}")
print(f"stack2.isFull() -> {stack2.isFull()}")
print(f"pop() -> {stack2.pop()}, stack = {stack2}")
print(f"pop() -> {stack2.pop()}, stack = {stack2}")
print(f"top() -> {stack2.top()}")
```

## Bài 5 - RESTful API quản lý thư viện bằng Flask

API quản lý danh sách sách với các chức năng: thêm sách mới, lấy danh sách sách, cập nhật sách, xóa sách. Mỗi sách có định dạng `{"id": value1, "title": value2, "author": value3}`.

Cell demo sử dụng `app.test_client()` để gọi API và in kết quả ngay trong notebook.

```python
from flask import Flask, jsonify, request


def create_library_app():
    app = Flask(__name__)
    books = [
        {"id": 1, "title": "Python for Engineers", "author": "Course Team"},
        {"id": 2, "title": "Clean Code", "author": "Robert C. Martin"},
    ]
    next_id = {"value": 3}

    def find_book(book_id):
        return next((book for book in books if book["id"] == book_id), None)

    @app.get("/books")
    def get_books():
        return jsonify(books), 200

    @app.post("/books")
    def add_book():
        data = request.get_json(silent=True) or {}
        title = data.get("title")
        author = data.get("author")
        if not title or not author:
            return jsonify({"error": "title and author are required"}), 400

        book = {"id": next_id["value"], "title": title, "author": author}
        next_id["value"] += 1
        books.append(book)
        return jsonify(book), 201

    @app.put("/books/<int:book_id>")
    def update_book(book_id):
        book = find_book(book_id)
        if book is None:
            return jsonify({"error": "book not found"}), 404

        data = request.get_json(silent=True) or {}
        if "title" in data:
            book["title"] = data["title"]
        if "author" in data:
            book["author"] = data["author"]
        return jsonify(book), 200

    @app.delete("/books/<int:book_id>")
    def delete_book(book_id):
        book = find_book(book_id)
        if book is None:
            return jsonify({"error": "book not found"}), 404

        books.remove(book)
        return jsonify({"message": "book deleted", "book": book}), 200

    return app


app = create_library_app()

with app.test_client() as client:
    response = client.get("/books")
    print("GET /books")
    print(response.status_code, response.get_json())

    response = client.post("/books", json={"title": "Flask API", "author": "Pallets"})
    created_book = response.get_json()
    print("\nPOST /books")
    print(response.status_code, created_book)

    response = client.put(f"/books/{created_book['id']}", json={"title": "Flask RESTful API"})
    print("\nPUT /books/<id>")
    print(response.status_code, response.get_json())

    response = client.delete(f"/books/{created_book['id']}")
    print("\nDELETE /books/<id>")
    print(response.status_code, response.get_json())

    response = client.get("/books")
    print("\nGET /books sau khi xóa")
    print(response.status_code, response.get_json())
```
