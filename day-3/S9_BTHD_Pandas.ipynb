# Hàm apply()

```python
import pandas as pd
import numpy as np

def Tang(x):
    return x if x > 500 else x + 1000

chi_so = ["Accounting", "Engineering", "Computer Science", "Mechanic"]
gia_tri = [300, 310, 311, 600]

S = pd.Series(gia_tri, index = chi_so) # dao lại bị sai

print(S.apply(Tang))
```

```python
import pandas as pd

# Tạo một DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}
df = pd.DataFrame(data)

print(df)

# Hàm để áp dụng
def sum_row(row):
    return row.sum()

# Áp dụng hàm `sum_row` dọc theo mỗi hàng (axis=1)
result = df.apply(sum_row, axis=1)
print(result)

# Áp dụng hàm `sum_row` dọc theo mỗi cột (axis=0)
result = df.apply(sum_row, axis=0)
print(result)
```

# Them 1 hang ở cuối cùng

```python
import pandas as pd

df = pd.read_csv("brics.csv")
print(df)


# Them dòng
new_row = pd.DataFrame({"Unnamed: 0":"VN", "country": "Vietnam", "population": [80.0], "area": 1880000.0, "capital":"hanoi"})

df_new = pd.concat([df,new_row], ignore_index=True)
print(df_new) 

# Lưu DataFrame mới vào file CSV
df_new.to_csv("brics_updated.csv", index=False)
de_new = pd.read_csv("brics_updated.csv", index_col=0)
print(de_new)

```

# Thay đổi tên nhãn ở 1 hàng bất kỳ

```python
import pandas as pd

df = pd.read_csv("data/brics.csv")
print(df)
print(df.describe())  # thống kê mô tả
print(df.info())  # thông tin về DataFrame

#Them côt
new_value = df["population"]/df["area"]*10000000
df.insert(1,"density",new_value)
df.iloc[3,2] = "Czech"    # đổi value
df = df.rename(index={0: 'test1', 1: "test2"})  # đổi tên cột
print(df)
```

```python
import pandas as pd

# Tạo một DataFrame ví dụ
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)

# In ra tên cột hiện tại
print("Tên cột hiện tại:")
print(df)

# Đổi tên cột 'Age' thành 'Years' sử dụng .rename()
df = df.rename(columns={'Age': 'Years'})

# In ra DataFrame sau khi đổi tên cột
print("\nTên cột sau khi đổi:")
print(df)
```
