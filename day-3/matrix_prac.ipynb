Tạo 1 ma trận 4x4 toàn False value

```python
# Import Libs here
import numpy as np
```

```python
# 4x4 matrix
a = np.array([
  [False, False, False, False],
  [False, False, False, False],
  [False, False, False, False],
  [False, False, False, False]
])

print(f"4x4 Matrix of False values: \n {a}")
```

```python
a = np.array([])
for i in range(100):
    a = np.append(a, i)

# Extract only odd values to new array
np_a = a[a % 2 == 1]

print(f"List of values: {a} \n")
print(f"Odd values: {np_a}")
```

```python
# Short-hand code of defining list of n values
org_a = np.arange(1, 20)
odd_a = np.copy(org_a)

# Replace only odd values with -1
odd_a[odd_a % 2 == 1] = -1

print(f"List of original values: {org_a} \n")
print(f"List of replaced values: {odd_a}")
```

```python
# Short-hand code of defining list of n - random values with numpy
list_a = np.random.randint(1, 30, size=20)
list_b = np.random.randint(1, 20, size=20)

# List c = concatenation of list a and list b
list_c = np.concatenate((list_a, list_b))
print(f"List A: {list_a} \n")
print(f"List B: {list_b} \n")

print(f"List C: {list_c} \n")
```

```python
# Short-hand code of defining list of n - random values with numpy
list_a = np.random.randint(1, 30, size=20)
list_b = np.random.randint(1, 20, size=20)

# List c = concatenation of list a and list b
list_c = np.array(
    [(list_a[i], list_b[i]) for i in range(len(list_a))]
)
print(f"List A: {list_a} \n")
print(f"List B: {list_b} \n")

print(f"List C: {list_c} \n")
```

```python
list_a = np.random.randint(1, 30, size=20)

list_c = list_a[list_a % 2 == 1]

print(f"List A: {list_a}\n")
print(f"List C: {list_c}")
```
